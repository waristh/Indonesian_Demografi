```python
df.drop(columns=['Area','Provinsi','Kota_Kabupaten','Regional','PDRB_Per_Kapita'],inplace=True)
```

# Memisah antara features & target

```python

X= df.drop("Indeks_Pembangunan_Manusia",axis=1)
y= df["Indeks_Pembangunan_Manusia"]
```

# Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler  
#Define MinMaxScaler as scaler  
scaler = MinMaxScaler(feature_range=(0,1))  
#Apply fit_transfrom to scale selected feature  
X = scaler.fit_transform(X)
#Cheking min and max value of the scaling_column
# print(pd.DataFrame(X).describe().T[['min','max']])
X = np.array(X)
X
```

# Melatih dengan beberapa Model

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
# checking the shapes  
print('Shape of X_train :', X_train.shape)
print('Shape of y_train :', y_train.shape)
print('Shape of X_test :', X_test.shape)
print('Shape of y_test :', y_test.shape)

##import regressor from Scikit-Learn
from  sklearn.linear_model import LinearRegression 
# Call the regressor
reg = LinearRegression()
# Fit the regressor to the training data  
reg = reg.fit(X_train, y_train)
# Apply the regressor/model to the test data  
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  
import numpy as np
import matplotlib.pyplot as plt 

#Calculating MSE, lower the value better it is. 0 means perfect prediction
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error of testing set:', mse)
#Calculating MAE
mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute error of testing set:', mae)
#Calculating RMSE
rmse = np.sqrt(mse)
print('Root Mean Squared Error of testing set:', rmse)
#Calculating r2_score
r2_score = r2_score(y_test, y_pred)
print('r2_score of testing set:', r2_score)

#Plotting y_test dan y_pred
plt.scatter(y_test, y_pred, c = 'green')
plt.xlabel('Indeks_Pembangunan_Manusia Actual')
plt.ylabel('Predicted value')
plt.title('True value vs predicted value : Linear Regression')
plt.show()

pd.DataFrame({'Actual':y_test,'Predict':y_pred})
```

```python
from sklearn.svm import SVR 
# Call the regressor
SVRegression = SVR()
# Fit the regressor to the training data  
model_svr = SVRegression.fit(X_train, y_train)
# Apply the regressor/model to the test data  
y_pred_SVR = model_svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  
import numpy as np
import matplotlib.pyplot as plt 

#Calculating MSE, lower the value better it is. 0 means perfect prediction
mse = mean_squared_error(y_test, y_pred_SVR)
print('Mean squared error of testing set:', mse)
#Calculating MAE
mae = mean_absolute_error(y_test, y_pred_SVR)
print('Mean absolute error of testing set:', mae)
#Calculating RMSE
rmse = np.sqrt(mse)
print('Root Mean Squared Error of testing set:', rmse)
#Calculating r2_score
r2_score = r2_score(y_test, y_pred_SVR)
print('r2_score of testing set:', r2_score)

#Plotting y_test dan y_pred
plt.scatter(y_test, y_pred_SVR, c = 'green')
plt.xlabel('Indeks_Pembangunan_Manusia Actual')
plt.ylabel('Predicted value')
plt.title('True value vs predicted value : SVR')
plt.show()

pd.DataFrame({'Actual':y_test,'Predict':y_pred_SVR})
```
